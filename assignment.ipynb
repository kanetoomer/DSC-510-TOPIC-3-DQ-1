```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Creating random data
np.random.seed(1)
returns = np.random.normal(0, 0.01, 1000)  # Mean = 0, Std = 0.01

# This is my DataFrame
returns_df = pd.DataFrame(returns, columns=['Returns'])

# I am calculating the VaR at 95% confidence level
confidence_level = 0.95
VaR = np.percentile(returns, (1 - confidence_level) * 100)

# This is my visual aid 
plt.hist(returns, bins=50, alpha=0.75, edgecolor='black')
plt.axvline(x=VaR, color='r', linestyle='--', label=f'VaR (95%): {VaR:.4f}')
plt.xlabel('Returns')
plt.ylabel('Frequency')
plt.title('Distribution of Returns')
plt.legend()
plt.show()

VaR
```
